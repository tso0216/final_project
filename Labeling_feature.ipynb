```python
import os
import kagglehub

path = kagglehub.dataset_download("brendanartley/cartoon-faces-googles-cartoon-set")
print("Path to dataset files:", path)
```

```python
path = os.path.join("/root/.cache/kagglehub/datasets/brendanartley/cartoon-faces-googles-cartoon-set/versions/1",'cartoonset100k_jpg/0')
path
```

```python
img = []
for root, dirs, files in os.walk(path):
    for f in files:
          full_path = os.path.join(root, f)
          img.append(full_path)
img
```

# labeling

```python
from pigeon import annotate
from IPython.display import display, Image
import pandas as pd

sub_img = img[1300:1550]
annotations = annotate(
    sub_img,
    options=['1','0'],
    display_fn=lambda filename: display(Image(filename))
)


df = pd.DataFrame(annotations, columns=['filepath', 'label'])
mapping = {'1': 1, '0': 0}
df['label'] = df['label'].map(mapping)
df.head()
```

```python
df.to_csv('gender.csv', index=False)
```
